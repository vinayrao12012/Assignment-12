## Q2

A scatter plot is a type of data visualization that is commonly used to display the relationship and distribution of data points in two continuous variables. It is particularly useful for identifying patterns, trends, clusters, or outliers in data. 

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Scatter plot")
plt.show()
```

## Q3

The subplot() function in Matplotlib is used to create a grid of subplots within a single figure, allowing you to display multiple plots or charts in a structured layout. This is useful when you want to compare different datasets, variables, or aspects of your data in a single figure.

```python
import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3= np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
plt.figure(figsize=(8,8))

plt.subplot(2,2,1)
plt.plot(x1,y1,color='#D36633')
plt.title("Subplot 1")

plt.subplot(2,2,2)
plt.plot(x2,y2,color="#CBE555")
plt.title("Subplot 2")

plt.subplot(2,2,3)
plt.plot(x3,y3,color="#55E5B3")
plt.title("Subplot 3")

plt.subplot(2,2,4)
plt.plot(x4,y4,color="#55BEE5")
plt.title("Subplot 4")

plt.tight_layout()
plt.show()
```

## Q4

A bar plot, also known as a bar chart or bar graph, is a type of data visualization that represents categorical data with rectangular bars. Each bar's length or height is proportional to the value it represents, making it a useful tool for visualizing and comparing discrete data .

why it is used?

1. Bar plots are used to compare the values or frequencies of different categories or groups.
2. Bar plots are useful when you want to compare data across discrete groups, such as comparing sales by region, student performance by grade, or survey responses by age group.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit)
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("Company's Profit")
plt.show()
```

## Q5

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It displays the summary statistics of a dataset, providing information about its central tendency, spread, and the presence of outliers.

why it is used?

1. Box plots help visualize the distribution of data and provide a quick overview of its characteristics, such as skewness, central tendency, and variability.
2. Identifying outliers:  Any data points beyond the whiskers are considered outliers, making it easy to spot them.


```python
import numpy as np
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.xlabel("Categories 1")
plt.ylabel("Values 1")
plt.title("Box Plot 1")
plt.boxplot(box1)
plt.show()
```

```python
plt.xlabel("Categories 2")
plt.ylabel("Values 2")
plt.title("Box Plot 2")
plt.boxplot(box2)
plt.show()
```
