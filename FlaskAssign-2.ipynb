## Q1

1. GET Method : The GET method is used to request data from a server. It appends the data as parameters in the URL's query string.

                 The data sent through a GET request is visible in the URL, which can be seen in the browser's address bar and in browser history.
                 
                 GET requests are generally used for retrieving data, like fetching a web page, an image, or data for display.
                 
                 
2. POST Method : Data sent through a POST request is included in the request body rather than in the URL.  

                 POST requests are not visible in the URL and are not cached by browsers, making them more suitable for sending sensitive or large amounts of data.
                               
                  POST requests are used for submitting forms, uploading files, and performing actions that change server state.

## Q5

```python
from flask import Flask, request,jsonify

app = Flask(__name__)

@app.route('/postman_action',methods=['POST'])
def math_ops():
    if (request.method == 'POST'):
        ops = request.json['operation']
        num1 = int(request.json['num1'])
        num2 = int(request.json['num2'])

        if ops == "add":
            r = num1+num2
            result = "The sum of " + str(num1) + ' and ' + str(num2) + " is " + str(r)

        if ops == "subtract":
            r = num1-num2
            result = "The sum of " + str(num1) + ' and ' + str(num2) + " is " + str(r)

        if ops == "multiply":
            r = num1*num2
            result = "The sum of " + str(num1) + ' and ' + str(num2) + " is " + str(r)

        if ops == "divide":
            r = num1/num2
            result = "The sum of " + str(num1) + ' and ' + str(num2) + " is " + str(r) 

        return jsonify(result) 


if __name__ == '__main__':
    app.run(host="0.0.0.0")    
```
