## Q1

Flask is a micro web framework for Python that is designed to be lightweight, flexible, and easy to use. It's commonly used to build web applications, APIs, and other web services. Flask provides the essentials for building web applications while allowing developers to choose and integrate additional components as needed.

Advantages of flask framework:

1. Lightweight and Flexible: Flask's minimalistic approach allows developers to choose and integrate only the components they need, making it flexible and well-suited for various project sizes and requirements.

2. Easy to Learn: Flask's simplicity and intuitive design make it relatively easy for beginners to learn and get started with web development.

3. Deployment Options: Flask applications can be deployed on various hosting platforms and cloud services, making it convenient to launch web applications to a broader audience.

## Q2

```python
from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "Hello, World!"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```
