## Q1

1. Histograms: Histograms are used to visualize the distribution of a single variable (univariate data). They show the frequency or count of data points within specified bins or intervals.
2. Scatter Plot:  Scatter plots are used to display the relationship between two numerical variables (bivariate data). They help identify patterns, trends, and correlations in the data.
3. Relplot: relplot is used for creating scatter plots and line plots to visualize the relationship between two numeric variables. It is a versatile function that can display the relationship between variables in different ways, including scatter plots, line plots, and more.
4. jointplot : jointplot is used for visualizing the relationship between two numeric variables and displaying their univariate distributions. It combines a scatter plot at its center with histograms  along the axes to show both the distribution of each variable and their joint distribution.
5. Pairplot: pairplot is used to create a matrix of scatter plots for pairwise relationships between multiple variables. It's particularly useful when you have several numeric variables, and you want to quickly visualize their interactions and correlations.

## Q2

```python
import seaborn as sns
```

```python
Fmri = sns.load_dataset("fmri")
```

```python
Fmri.head()
```

```python
sns.lineplot(x="timepoint",y="signal",data = Fmri)
```

```python
sns.lineplot(x="timepoint",y="signal",data = Fmri, hue='event')
```

```python
sns.lineplot(x="timepoint",y="signal",data = Fmri, hue='event',style='region')
```
